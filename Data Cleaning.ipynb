# Salary Estimates



```python
#we can see that there are some rows with -1 hence removing - -1
df = df[df["Salary Estimate"] != "-1"]

salary = df["Salary Estimate"].apply(lambda x : x.split("(")[0])

#creating new column for hourly salary from salary estimate column as its is distinct.
df["hourly"] = df["Salary Estimate"].apply( lambda x : 1 if "per hour" in x.lower() else 0 )

#creating new column for employer_provided salary from salary estimate column as its is distinct.
df["employer_provided"] = df["Salary Estimate"].apply(lambda x : 1 if "employer provided salary:" in x.lower() else 0)
```

```python
df.head()
```

```python
replace_kd = salary.apply(lambda x : x.replace("$"," ").replace("K"," "))

replace_perhour_Employer_sal =replace_kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
replace_perhour_Employer_sal
```

```python
df["min_salary"] = replace_perhour_Employer_sal.apply(lambda x: int(x.split('-')[0]))
df["max_salary"] = replace_perhour_Employer_sal.apply(lambda x: int(x.split('-')[1]))
```

```python
df["Avg_salary"] = (df.min_salary + df.max_salary)/2
```

```python
df.head()
```

# Company name text only

```python
df["Company Name_text"] = df.apply(lambda x : x["Company Name"] if x["Rating"] < 0 else x["Company Name"][:-3], axis = 1)
```

```python
df.head()
```

# Location

```python
#lets get state from location table
df["job_state"] = df["Location"].apply(lambda x : x.split(",")[1])
df.head()
```

```python
#column for same QS and job location
df["same_state_HQ and job location"] = df.apply(lambda x : 1 if x["Location"] == x["Headquarters"] else 0, axis = 1)
```

```python
#just to check the how many jobs are there in each state of USA
df.job_state.value_counts()
```

# Founded

```python
#to get the age of the company to know how old is the company

df["age_of_company"] = df["Founded"].apply(lambda x : x if x < 1 else 2022 - x) 
```
